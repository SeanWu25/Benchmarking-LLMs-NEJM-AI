```python
import pandas as pd
import warnings
import re

def count_matches(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Initialize count
    count = 0

    # Create a new DataFrame to store the mismatched rows
    matched_rows = pd.DataFrame(columns=df.columns)

    # Initialize total questions
    total_questions = len(df)

    # Define the regular expressions to match
    regex_patterns = [
        r'Answer: ([A-E])\.',
        r'answer: ([1-5])\.',
        r'answer: ([A-E])$',
        r"Answer: '([1-5])\.",
        r"Answer: '([A-E])\)",
        r"correct option is ([1-5])\.",
        r"correct option is ([A-E])\.",
        r'Explanation: ([1-5]) is correct',
        r'Explanation: ([A-E]) is correct',
        r'Answer: Option ([1-5]),',
        r'Answer: Option ([A-E]),',
        r'Choice ([1-5])',
        r'Choice ([A-E])',
        r"^([A-E])$",
        r'^\(([A-E])\)',
        r"would be option ([1-5]),",
        r"would be option ([A-E]),"
    ]

    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        # Extract values from column 1 and column 2
        value1 = str(row[1])
        value2 = str(row[2])

        value1 = value1.upper()

        if len(value1) > 1:
          continue

        # Check for matches
        for pattern in regex_patterns:
                match = re.search(pattern, value2)
                if match:
                    v = match.group(1)
                    if (value1[0] == 'A' and v == '1') or (value1[0] == 'B' and v == '2') or \
                       (value1[0] == 'C' and v == '3') or (value1[0] == 'D' and v == '4') or \
                       (value1[0] == 'E' and v == '5') or (v == value1[0]):
                        count += 1
                        with warnings.catch_warnings():
                            warnings.simplefilter("ignore")
                            matched_rows = matched_rows.append(row)
                        break

    # Save the mismatched rows to a CSV file
    matched_rows.to_csv('vicuna_only_matched_rows.csv', index=False)

    return count, total_questions

# Specify the input CSV file path
csv_file = '/content/vicuna_End-Stage Kidney Disease_rows.csv'

# Call the count_matches function
total_matches, total_questions = count_matches(csv_file)
print('Total matches:', total_matches)
print('Total Questions:', total_questions)
print('Total matches percentage:', total_matches/total_questions)
```
