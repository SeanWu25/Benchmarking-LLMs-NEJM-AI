```python
import json
import replicate
idlist = []
predictedanswer = []
groundtruth = []
with open("C:/Users/Admin/Downloads/Second_Half_Nephrology.json") as file:
    data = json.load(file)

for item in data:
  #item is the ID
  context = "Context: " + data[item]['context']
  question = "Question: " + data[item]['question']
  choices = "Choices (Pick one): " + str(data[item]['choices'])
  answer = data[item]['answer']

  idlist.append(item)
  groundtruth.append(answer)

  combined_string = context + "\n" + question + "\n" + str(choices)
  #print(combined_string)
  output = replicate.run(
    "meta/llama-2-70b-chat:2c1608e18606fad2812020dc541930f2d0495ce32eee50074220b87300bc16e1",
    input={"prompt": combined_string})

  computeranswer = ""
  for item in output:
      computeranswer += item


  print("THE COMPUTER ANSWER FOR ID: " + str(item) + " is " + computeranswer)
  print("---------------------------------------------------------------------------------------------------")
  predictedanswer.append(computeranswer)



```

```python
combined_string  = 'Context: Question: In addition to patients with acute leukemia, patients with which ONE of the following diagnosesis LEAST likely to receive inpatient palliative care services based on data from the United States National Inpatient Sample Data? Choices (Pick one): [A. Acute respiratory distress syndrome, B. Cardiogenic shock, C. End-stage liver disease with cirrhosis, D. AKI requiring dialysis]'

output = replicate.run(
"meta/llama-2-70b-chat:2c1608e18606fad2812020dc541930f2d0495ce32eee50074220b87300bc16e1",
input={"prompt": combined_string})

computeranswer = ""
for item in output:
  computeranswer += item
```

```python
import csv

header = ['ID', 'Ground_Truth', 'LLama70b_Output']

data = zip(idlist, groundtruth, predictedanswer)

filename = 'C:/Users/Admin/Desktop/LLama70b/second_half_LLama70b.csv'

with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(header)
    writer.writerows(data)

print(f"Data has been written to {filename} successfully.")
```
