```python
from transformers import LlamaTokenizer, LlamaForCausalLM, GenerationConfig, pipeline
import torch
import textwrap
```

```python
tokenizer = LlamaTokenizer.from_pretrained("samwit/koala-7b")

base_model = LlamaForCausalLM.from_pretrained(
    "samwit/koala-7b",
    load_in_8bit=True,
    device_map='auto',
)
```

```python
!nvidia-smi
```

```python
import textwrap

pipe = pipeline(
    "text-generation",
    model=base_model,
    tokenizer=tokenizer,
    max_length=700,
    temperature=0.7,
    top_p=0.95,
    repetition_penalty=1.15
)

def wrap_text_preserve_newlines(text, width=110):
    # Split the input text into lines based on newline characters
    lines = text.split('\n')

    # Wrap each line individually
    wrapped_lines = [textwrap.fill(line, width=width) for line in lines]

    # Join the wrapped lines back together using newline characters
    wrapped_text = '\n'.join(wrapped_lines)

    return wrapped_text
```

```python
import json
idlist = []
predictedanswer = []
groundtruth = []
with open('/content/data.json') as file:
    data = json.load(file)

for item in data:
  #item is the ID
  context = "Context: " + data[item]['context']
  question = "Question: " + data[item]['question']
  choices = "Choices (Pick one): " + str(data[item]['choices'])
  answer = data[item]['answer']

  idlist.append(item)
  groundtruth.append(answer)

  combined_string = context + "\n" + question + "\n" + str(choices)
  output = pipe(combined_string)

  computeranswer = output[0]['generated_text']

  print("THE COMPUTER ANSWER FOR ID: " + str(item) + " is " + computeranswer)
  print("---------------------------------------------------------------------------------------------------")
  predictedanswer.append(computeranswer)




```

```python
import csv

header = ['ID', 'Ground_Truth', 'Koala_Output']

data = zip(idlist, groundtruth, predictedanswer)

filename = 'final_koala_section.csv'

with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(header)
    writer.writerows(data)

print(f"Data has been written to {filename} successfully.")
```
