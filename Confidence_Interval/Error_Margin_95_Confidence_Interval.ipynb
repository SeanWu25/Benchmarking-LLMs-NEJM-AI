```python
#(p - 1.96*sqrt(p(1-p)/n) , p + 1.96*sqrt(p(1-p)/n) )

import math

def confidence_interval(p, n = 858):
  lower = p - 1.96* math.sqrt(p*(1-p)/n)
  upper = p + 1.96*math.sqrt(p*(1-p)/n)
  return (lower,upper)


print("GPT-4 95% Confidence Interval")
print(confidence_interval(73.3/100))
print("")

print("Claude-2 95% Confidence Interval")
print(confidence_interval(54.4/100))
print("")

print("Vicuna 95% Confidence Interval")
print(confidence_interval(25.5/100))
print("")

print("Orca 95% Confidence Interval")
print(confidence_interval(17.1/100))
print("")

print("Falcon 95% Confidence Interval")
print(confidence_interval(18.1/100))
print("")

print("Koala 95% Confidence Interval")
print(confidence_interval(23.8/100))
print("")

print("Llama 95% Confidence Interval")
print(confidence_interval(30.7/100))
print("")

```
